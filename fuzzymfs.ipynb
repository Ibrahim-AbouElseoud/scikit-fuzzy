```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
# Universe variable
x = np.arange(0, 10.1, 0.1)

# Fuzzy membership functions
tri = fuzz.trimf(x, [1, 7, 9])
gauss = fuzz.gaussmf(x, 7, 3)

trap = fuzz.trapmf(x, [1, 2, 4, 7])
gauss2 = fuzz.gauss2mf(x, 2, 0.5, 4, 1.2)
pimf = fuzz.pimf(x, 1, 2, 4, 7)

smf = fuzz.smf(x, 4, 7)
zmf = fuzz.zmf(x, 4, 7)
s_trap = fuzz.trapmf(x, [4, 7, 10, 10])
z_trap = fuzz.trapmf(x, [0, 0, 4, 7])

sig = fuzz.sigmf(x, 3, -2)
dsig = fuzz.dsigmf(x, 3, 2, 6, 4)
psig = fuzz.psigmf(x, 5.5, -3, 6, 4)

gbell = fuzz.gbellmf(x, 2, 4, 6)
piece = fuzz.piecemf(x, [1, 3, 10])
```

```python
# Generate plot
plt.figure()
gs = gridspec.GridSpec(3, 4)
axes = []

# Line 0
ax0 = plt.subplot(gs[0, 0])
ax0.plot(x, tri, 'r')
axes.append(ax0)

ax1 = plt.subplot(gs[0, 1])
ax1.plot(x, gauss, 'r')
axes.append(ax1)

ax2 = plt.subplot(gs[0, 2])
ax2.plot(x, s_trap, 'r', x, z_trap, 'b')
axes.append(ax2)

ax3 = plt.subplot(gs[0, 3])
ax3.plot(x, smf, 'r', x, zmf, 'b')
axes.append(ax3)

# Line 1
ax4 = plt.subplot(gs[1, 0])
ax4.plot(x, trap, 'r')
axes.append(ax4)

ax5 = plt.subplot(gs[1, 1])
ax5.plot(x, gauss2, 'r')
axes.append(ax5)

ax6 = plt.subplot(gs[1, 2])
ax6.plot(x, pimf, 'r')
axes.append(ax6)

ax7 = plt.subplot(gs[1, 3])
ax7.plot(x, gbell, 'r')
axes.append(ax7)

# Line 2
ax8 = plt.subplot(gs[2, 0])
ax8.plot(x, piece, 'r')
plt.ylim(-0.05, 1.05)
axes.append(ax8)

ax9 = plt.subplot(gs[2, 1])
ax9.plot(x, sig, 'r')
axes.append(ax9)

ax10 = plt.subplot(gs[2, 2])
ax10.plot(x, dsig, 'r')
axes.append(ax10)

ax11 = plt.subplot(gs[2, 3])
ax11.plot(x, psig, 'r')
axes.append(ax11)

# General plot spacing
plt.subplots_adjust(left=0.075, right=.95, bottom=0.05, top=0.95,
                    hspace=0.075, wspace=0.075)

# Turn off axis labeling except for outside axes
for i, ax in enumerate(axes):
    ax.set_ylim(-0.05, 1.05)
    
    if i not in [0, 4, 8]:
        ax.set_yticklabels([])
    else:
        ax.set_ylabel('Fuzzy membership')
        
    if i not in [8, 9, 10, 11]:
        ax.set_xticklabels([])
    else:
        ax.set_xlabel('Universe variable')

plt.show()
```
